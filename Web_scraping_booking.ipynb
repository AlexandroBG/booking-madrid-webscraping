````python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# Ruta correcta de ChromeDriver en Mac (ajusta si es necesario)
chrome_driver_path = r"/usr/local/bin/chromedriver"

# Configurar Selenium
service = Service(chrome_driver_path)
options = Options()
options.add_argument("--start-maximized")  # Maximizar ventana

# Inicializar el navegador
driver = webdriver.Chrome(service=service, options=options)

# URL de Booking con hoteles en Madrid
url = "https://www.booking.com/"
driver.get(url)

# Esperar 5 segundos para cargar la página
time.sleep(5)

print(" Página de Booking cargada correctamente")

````

## <span style="color:#003580">Seleccionar Número de Huéspedes y Habitaciones </span> <a class="anchor" id="1"></a>
```python

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

#  DATOS DE ENTRADA (solo 1 niño)
num_adults = 2   # Número de adultos
num_kids = 1     # Un solo niño
kid_age = 5      # Edad del niño
num_rooms = 1    # Número de habitaciones

try:
    # 1 ABRIR EL MENÚ DE HUÉSPEDES
    guests_button = driver.find_element(By.XPATH, '//button[contains(@data-testid, "occupancy-config")]')
    guests_button.click()
    time.sleep(2)

    # 2 SELECCIONAR NÚMERO DE ADULTOS
    adults_input = driver.find_element(By.ID, "group_adults")
    current_adults = int(adults_input.get_attribute("value"))

    if current_adults < num_adults:
        increase_adults = driver.find_element(By.XPATH, '//input[@id="group_adults"]/following-sibling::div/button[2]')
        for _ in range(num_adults - current_adults):
            increase_adults.click()
            time.sleep(1)
    elif current_adults > num_adults:
        decrease_adults = driver.find_element(By.XPATH, '//input[@id="group_adults"]/following-sibling::div/button[1]')
        for _ in range(current_adults - num_adults):
            decrease_adults.click()
            time.sleep(1)

    # 3 SELECCIONAR 1 NIÑO
    kids_input = driver.find_element(By.ID, "group_children")
    current_kids = int(kids_input.get_attribute("value"))

    if current_kids < num_kids:
        increase_kids = driver.find_element(By.XPATH, '//input[@id="group_children"]/following-sibling::div/button[2]')
        increase_kids.click()
        time.sleep(2)  # Esperar a que aparezca el selector de edad

    # 4 SELECCIONAR EDAD DEL NIÑO
    age_select_xpath = '//select[@name="age"]'
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, age_select_xpath)))
    age_selector = driver.find_element(By.XPATH, age_select_xpath)
    select = Select(age_selector)
    select.select_by_value(str(kid_age))  # Seleccionar la edad del niño
    time.sleep(1)

    # 5 SELECCIONAR NÚMERO DE HABITACIONES
    rooms_input = driver.find_element(By.ID, "no_rooms")
    current_rooms = int(rooms_input.get_attribute("value"))

    if current_rooms < num_rooms:
        increase_rooms = driver.find_element(By.XPATH, '//input[@id="no_rooms"]/following-sibling::div/button[2]')
        increase_rooms.click()
        time.sleep(1)
    elif current_rooms > num_rooms:
        decrease_rooms = driver.find_element(By.XPATH, '//input[@id="no_rooms"]/following-sibling::div/button[1]')
        decrease_rooms.click()
        time.sleep(1)

    # 6 CONFIRMAR SELECCIÓN CON EL BOTÓN "LISTO"
    confirm_button = driver.find_element(By.XPATH, '//button[span[text()="Listo"]]')
    confirm_button.click()
    print(f" Selección confirmada: {num_adults} adultos, {num_kids} niño ({kid_age} años), {num_rooms} habitación.")
    time.sleep(2)

except Exception as e:
    print(f" Error al seleccionar huéspedes: {e}")

````

## <span style="color:#003580">Extraer Información de Alojamientos </span> <a class="anchor" id="1"></a>
```python

### <span style="color:#0071c2">Cargar todos los alojamiento</span> <a class="anchor" id="1.1"></a>
```python


````python
# 1 Cargar todos los alojamientos (repetir hasta que ya no haya más resultados disponibles)
while True:
    try:
        # Esperar que el botón "Cargar más resultados" sea clickeable
        load_more_button_xpath = '//span[contains(text(), "Cargar más resultados")]'
        load_more_button = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, load_more_button_xpath)))

        # Hacer clic en el botón
        load_more_button.click()
        print("📥 Se han cargado más resultados...")

        # Esperar unos segundos para que se actualice la página
        time.sleep(3)

    except:
        print(" No hay más resultados para cargar.")
        break  # Salir del bucle si ya no hay más hoteles disponibles

# 2 Extraer nombres de hoteles (AHORA se cargaron todos los resultados)
hotel_elements = driver.find_elements(By.XPATH, '//div[@data-testid="title"]')
hotel_names = [hotel.text for hotel in hotel_elements]

print(f" Hoteles encontrados: {len(hotel_names)}")
print("Ejemplo de hoteles:", hotel_names[:10])
````

### <span style="color:#0071c2">Extraer nombres de alojamientos</span> <a class="anchor" id="1.1"></a>
```python

````python
# Extraer nombres de alojamientos (no solo hoteles)
alojamiento_elements = driver.find_elements(By.XPATH, '//div[@data-testid="title"]')
alojamientos = [alojamiento.text for alojamiento in alojamiento_elements]

print(f" Alojamientos encontrados: {len(alojamientos)}")
print("Ejemplo de alojamientos:", alojamientos[:10])

````

### <span style="color:#0071c2">Extraer Precios </span> <a class="anchor" id="1.1"></a>
```python

````python
# Extraer precios de los Alojamientos
price_elements = driver.find_elements(By.XPATH, '//span[@data-testid="price-and-discounted-price"]')
prices = [price.text for price in price_elements]

print(f" Precios encontrados: {len(prices)}")
print("Ejemplo de precios:", prices[:10])
````

### <span style="color:#0071c2">Extraer Puntuación Especifica</span> <a class="anchor" id="1.1"></a>
```python

````python
# Encontrar todos los bloques de alojamientos en la página
hotel_blocks = driver.find_elements(By.XPATH, '//div[@data-testid="property-card"]')

# Lista para almacenar puntuaciones correctamente alineadas
ratings = []

# Recorrer cada alojamiento y extraer su puntuación específica
for hotel in hotel_blocks:
    try:
        # Buscar la puntuación dentro del bloque del hotel
        rating_element = hotel.find_element(By.XPATH, './/div[contains(@class, "abf093bdfe d86cee9b25")]')
        rating_text = rating_element.text.strip()

        # Extraer solo el número (ejemplo: "Puntuación: 8,8 " → "8,8")
        rating = rating_text.split()[-1].replace(',', '.')  # Reemplazar coma por punto
    except:
        rating = "Sin puntuación"

    # Agregar la puntuación a la lista
    ratings.append(rating)

# Verificar la cantidad de puntuaciones extraídas
print(f" Puntuaciones encontradas: {len(ratings)}")
print("Ejemplo de puntuaciones:", ratings[:10])
````

### <span style="color:#0071c2">Extraer Ubicación</span> <a class="anchor" id="1.1"></a>
```python

````python
# Extraer ubicaciones
location_elements = driver.find_elements(By.XPATH, '//span[@data-testid="address"]')

# Obtener los textos de ubicación
locations = [location.text for location in location_elements]

print(f"📍 Ubicaciones encontradas: {len(locations)}")
print("Ejemplo de ubicaciones:", locations[:10])
````

### <span style="color:#0071c2">Extraer Reseñas</span> <a class="anchor" id="1.1"></a>
```python

````python
# Lista para almacenar reseñas correctamente alineadas
num_reviews = []

# Recorrer cada hotel en los bloques extraídos (para asegurar que la reseña corresponde al hotel correcto)
for hotel in hotel_blocks:
    try:
        # Buscar la reseña dentro del bloque del hotel
        review_element = hotel.find_element(By.XPATH, './/span[contains(@class, "f45d8e4c32 d935416c47")]')
        review_text = review_element.text.strip()

        # Extraer solo el número de comentarios
        review = review_text.replace(".", "").replace(" comentarios", "") + " comentarios"
    except:
        review = "No hay reseñas"

    # Agregar la reseña a la lista
    num_reviews.append(review)

# **Verificar si las reseñas están alineadas correctamente**
print(f" Reseñas encontradas: {len(num_reviews)}")
print(f" Hoteles encontrados: {len(hotel_blocks)}")

# **Asegurar que coincidan las longitudes**
if len(num_reviews) == len(hotel_blocks):
    print(" Las reseñas están alineadas correctamente con los hoteles.")
else:
    print(" ERROR: Las reseñas y los hoteles NO tienen el mismo tamaño.")

    # **Si hay un desfase, ajustar con "No hay reseñas"**
    while len(num_reviews) < len(hotel_blocks):
        num_reviews.append("No hay reseñas")
    while len(num_reviews) > len(hotel_blocks):
        num_reviews.pop()

    print(" Se ha ajustado la longitud de las reseñas para que coincida con los hoteles.")
    print(f"Nuevo total de reseñas: {len(num_reviews)}")
````

### <span style="color:#0071c2">Número de estrellas</span> <a class="anchor" id="1.1"></a>
```python

````python
# Encontrar todos los bloques de Alojamientos en la página
hotel_blocks = driver.find_elements(By.XPATH, '//div[@data-testid="property-card"]')

# Lista para almacenar estrellas correctamente alineadas
stars = []

# Recorrer cada hotel y extraer el número de estrellas
for hotel in hotel_blocks:
    try:
        # Buscar el contenedor de estrellas dentro del hotel
        stars_element = hotel.find_element(By.XPATH, './/div[@data-testid="rating-stars"]')

        # Contar la cantidad de iconos de estrellas dentro del contenedor
        num_stars = len(stars_element.find_elements(By.TAG_NAME, "svg"))
    except:
        num_stars = 0  # Si hay error o no se encuentran estrellas, asignar 0

    # Convertir 0 en "Sin estrellas"
    stars.append(num_stars if num_stars > 0 else "Sin estrellas")

# **Asegurar que la lista de estrellas tenga el mismo tamaño que la de hoteles**
stars = stars + ["Sin estrellas"] * (len(hotel_names) - len(stars))

# Verificar la cantidad de estrellas extraídas
print(f" Hoteles con estrellas detectadas: {len(stars)}")
print("Ejemplo de estrellas:", stars[:10])

````

````python
print(f" Alojamientos encontrados: {len(alojamientos)}")  
print(f" Precios encontrados: {len(prices)}")  
print(f" Puntuaciones encontradas: {len(ratings)}")  
print(f" Reseñas encontradas: {len(num_reviews)}")  
print(f" Ubicaciones encontradas: {len(locations)}")  
print(f" Estrellas detectadas: {len(stars)}")  

# Comprobación final
if len(set([len(alojamientos), len(prices), len(ratings), len(num_reviews), len(locations), len(stars)])) == 1:
    print(" Todos los datos están alineados correctamente.")
else:
    print(" ERROR: Las listas de datos no tienen la misma longitud. Revisa las extracciones.")
````

Los datos extraídos contenían ruido y formatos inconsistentes, por lo que se llevó a cabo un proceso de limpieza y normalización utilizando Pandas.

**Tareas realizadas:**
* Conversión de precios: Eliminación del símbolo € y transformación a valores numéricos.  
* Corrección de puntuaciones: Conversión de cadenas a valores flotantes.  
* Normalización de ubicaciones: Separación en Distrito y Ciudad.  
* Manejo de valores faltantes: Imputación con valores por defecto o mediana.

A continuación, se presenta el código utilizado para limpiar y estructurar los datos.



````python
import pandas as pd
import numpy as np
````

````python
# Cargar el dataset extraído de Booking
df = pd.read_csv("alojamientos_booking.csv")

# Mostrar las primeras filas
print("🔍 Primeras filas del DataFrame:")
print(df.head())
````

````python
# Ver información general del DataFrame
print(" Información inicial del DataFrame:")
print(df.info())

# Revisar valores únicos en cada columna
for col in df.columns:
    print(f"\nValores únicos en {col}: {df[col].unique()[:10]}")  # Mostrar solo los primeros 10 valores únicos
````

````python
# Contar valores nulos en cada columna
print("\n Valores nulos antes de la limpieza:")
print(df.isnull().sum())
````

````python
import pandas as pd

# Cargar los datos sin conversión automática
df = pd.read_csv("alojamientos_booking.csv", dtype=str)

# 1 **Eliminar símbolos y espacios en 'Precio'**
df["Precio"] = df["Precio"].str.replace("€", "").str.replace(".", "").str.replace(",", "").str.strip()

# 2 **Convertir a número**
df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")

# 3 **Corregir valores que están mal escalados**
df["Precio"] = df["Precio"].apply(lambda x: x / 100 if x > 10000 else x)

# **Verificar los primeros resultados**
print("\n Precios correctamente normalizados:")
print(df[['Alojamiento', 'Precio']].head(10))
````

````python
# Convertir puntuaciones a valores numéricos
df['Puntuación'] = pd.to_numeric(df['Puntuación'], errors='coerce')

# Reemplazar NaN por 0.0 sin 'chained assignment'
df.loc[df['Puntuación'].isna(), 'Puntuación'] = 0.0

# Mostrar primeros valores para verificar
print("\n Puntuaciones convertidas a valores numéricos:")
print(df[['Alojamiento', 'Puntuación']].head())
````

````python
# Eliminar el carácter "·" y la palabra "comentarios"
df['Reseñas'] = df['Reseñas'].astype(str).str.replace("·", "", regex=True)
df['Reseñas'] = df['Reseñas'].str.replace(" comentarios", "", regex=True)

# Convertir a números enteros, tratando valores nulos como 0
df['Reseñas'] = pd.to_numeric(df['Reseñas'], errors='coerce').fillna(0).astype(int)

# Mostrar primeros valores
print("\n Reseñas convertidas a valores numéricos:")
print(df[['Alojamiento', 'Reseñas']].head())
````

````python
# Separar la columna de Ubicación en Distrito y Ciudad
df[['Distrito', 'Ciudad']] = df['Ubicación'].str.split(',', expand=True)

# Limpiar espacios en blanco en ambas columnas
df['Distrito'] = df['Distrito'].str.strip()
df['Ciudad'] = df['Ciudad'].str.strip()

# Rellenar valores faltantes en Ciudad con "Madrid" si Distrito ya es "Madrid"
df['Ciudad'] = df['Ciudad'].fillna("Madrid")
df.loc[df['Distrito'] == "Madrid", 'Ciudad'] = "Madrid"

# Mostrar primeros valores
print("\n Ubicaciones normalizadas (Distrito y Ciudad corregidos):")
print(df[['Alojamiento', 'Distrito', 'Ciudad']].head(15))


````

````python
# Reemplazar "Sin estrellas" por 0 y convertir a número
df['Estrellas'] = df['Estrellas'].replace("Sin estrellas", 0).astype(float)

# Mostrar primeros valores
print("\n Estrellas normalizadas:")
print(df[['Alojamiento', 'Estrellas']].head())
````

````python
import pandas as pd

# Crear DataFrame con información de los valores faltantes
df_info = pd.DataFrame({
    "Columnas": df.columns,
    "Datos Faltantes": df.isnull().sum().values,
    "Tipo de Dato": df.dtypes.values
})

# Aplicar estilos para mejor visualización en Jupyter Notebook
df_info_styled = df_info.style.set_table_styles([
    {"selector": "thead th", "props": "background-color: #003580; color: white; font-weight: bold; text-align: center;"},
    {"selector": "tbody td", "props": "text-align: center;"}]
).set_caption(" Resumen de Datos Faltantes")

# Mostrar en Jupyter Notebook
display(df_info_styled)
````

````python
# Guardar los datos limpios en CSV y Excel
csv_filename = "alojamientos_booking_limpio.csv"
excel_filename = "alojamientos_booking_limpio.xlsx"

df.to_csv(csv_filename, index=False, encoding="utf-8")
df.to_excel(excel_filename, index=False)

# Mensaje de confirmación
print(f" Datos limpios guardados en:\n- {csv_filename}\n- {excel_filename}")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos limpios
df = pd.read_csv("alojamientos_booking_limpio.csv")

# Ver las primeras filas para asegurarnos de que los datos se cargaron correctamente
df.head()
````

````python
plt.figure(figsize=(10,5))
sns.histplot(df["Precio"], bins=50, kde=True, color="blue")
plt.title("Distribución de Precios de los Alojamientos")
plt.xlabel("Precio (€)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
````

</div>

**Interpretación:**  
El histograma de distribución de precios de los alojamientos muestra una gran variabilidad en los costos. Se observa que la mayoría de los alojamientos se concentran en un rango de precios entre 1000€ y 3000€, lo que indica que este es el segmento más accesible para los viajeros.  

Sin embargo, la distribución presenta una cola larga hacia valores más altos, lo que evidencia la existencia de alojamientos de lujo con precios significativamente elevados. Estos casos representan una minoría en la muestra, pero reflejan la diversidad de opciones en el mercado hotelero de Madrid.  



````python
 plt.figure(figsize=(10,6))
sns.boxplot(x=df["Precio"], color="tomato")

# Configurar título y etiquetas
plt.title(" Boxplot del Precio de los Alojamientos", fontsize=16, fontweight='bold', color='#333333')
plt.xlabel("Precio (€)", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
````

</div>

**Interpretación:**  
El boxplot complementa este análisis al revelar la presencia de outliers (valores atípicos) que superan los 6000€. Estos puntos corresponden a alojamientos exclusivos o de alta gama, reflejando la diversidad de opciones disponibles en la ciudad.  

Además, se observa que la mediana del precio se encuentra en 1992.50€, lo que representa el costo típico de los alojamientos en Madrid. Esto indica que, aunque existen alojamientos de lujo con precios elevados, la mayoría de los hospedajes se sitúan en un rango más accesible para los viajeros.  


````python
plt.figure(figsize=(10,5))
sns.histplot(df["Puntuación"], bins=20, kde=True, color="green")
plt.title("Distribución de Puntuaciones de los Alojamientos")
plt.xlabel("Puntuación")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
````

</div>

**Interpretación:**  
El histograma de puntuaciones muestra que la mayoría de los alojamientos en Madrid tienen valoraciones superiores a 7.0, con un pico en 8.0 y 9.0. Esto sugiere que, en general, los usuarios perciben positivamente los alojamientos en la ciudad.  

Sin embargo, también se observa un pequeño grupo de alojamientos con puntuaciones cercanas a 0 o 1, lo que indica la presencia de experiencias negativas por parte de algunos huéspedes. Esta dispersión en las valoraciones refleja la diversidad en la calidad del servicio y la satisfacción del cliente dentro del sector hotelero en Madrid.  


````python
plt.figure(figsize=(10,6))
sns.histplot(df["Reseñas"], bins=30, kde=True, color="royalblue", edgecolor="black")

# Configurar título y etiquetas
plt.title("Distribución de la Cantidad de Reseñas por Alojamiento", fontsize=16, fontweight='bold', color='#333333')
plt.xlabel("Número de Reseñas", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
````

**Interpretación:**  
Se analizó la cantidad de reseñas por alojamiento, observándose que una gran proporción de alojamientos tiene menos de 500 reseñas, mientras que un pequeño grupo supera las 10,000 opiniones.  

Este patrón sugiere que los alojamientos más populares concentran una mayor cantidad de reseñas, lo que indica una alta demanda y visibilidad en la plataforma. En contraste, la mayoría de los establecimientos tienen menor volumen de clientes o menor exposición, lo que podría afectar su competitividad en el mercado.  


````python
plt.figure(figsize=(10,5))
sns.boxplot(x="Estrellas", y="Precio", data=df, hue="Estrellas", palette="coolwarm", legend=False)
plt.title("Relación entre Precio y Número de Estrellas")
plt.xlabel("Número de Estrellas")
plt.ylabel("Precio (€)")
plt.grid(True)
plt.show()

````

</div>

**Interpretación:**  
El boxplot que relaciona el número de estrellas con el precio confirma que los alojamientos de mayor categoría tienden a ser más costosos.  

* Los hoteles de 5 estrellas presentan los precios más elevados, con una mediana superior a 4,000€.  
* Los alojamientos de 1 y 2 estrellas tienen precios considerablemente más bajos, con menor variabilidad en su rango de costos.  
* Los alojamientos sin clasificación de estrellas muestran una amplia dispersión de precios, lo que sugiere que muchos apartamentos turísticos y hostales no siguen una estructura tarifaria estándar, abarcando desde opciones económicas hasta alojamientos de lujo.  

Este análisis refleja la diversidad del mercado hotelero en Madrid, donde la categoría de estrellas no siempre es un factor determinante del precio final.

**Interpretación:**  
El gráfico de dispersión que relaciona la puntuación y el precio indica que no existe una correlación fuerte entre ambas variables.  

* Algunos de los alojamientos más costosos tienen altas puntuaciones, lo que podría reflejar servicios premium y mayor calidad percibida.  
* Sin embargo, también hay opciones económicas con valoraciones elevadas, lo que sugiere que los usuarios valoran otros aspectos además del precio, como la ubicación, la comodidad y la experiencia general.

Este análisis destaca la diversidad del mercado hotelero en Madrid, donde los precios no siempre reflejan la satisfacción del cliente.  


````python
### 6. Relación entre Número de Reseñas y Puntuación ###
plt.figure(figsize=(8,6))
sns.scatterplot(x="Reseñas", y="Puntuación", data=df, hue="Puntuación", palette="coolwarm", size="Reseñas", sizes=(10, 200), edgecolor="black", alpha=0.7)
plt.title("Relación entre Número de Reseñas y Puntuación", fontsize=16, fontweight='bold', color='#333333')
plt.xlabel("Número de Reseñas", fontsize=12)
plt.ylabel("Puntuación", fontsize=12)
plt.xscale("log")  # Escala logarítmica para mejor visualización
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
````

**Interpretación:**  
El gráfico de burbujas revela que los alojamientos con más reseñas tienden a mantener puntuaciones altas, lo que sugiere que los establecimientos con mayor volumen de clientes suelen ofrecer experiencias satisfactorias.

Este análisis destaca que la popularidad de un alojamiento no siempre garantiza una calificación perfecta, ya que la diversidad de experiencias de los huéspedes influye en la puntuación global.

````python
import matplotlib.pyplot as plt

# Contar la cantidad de alojamientos por número de estrellas
estrellas_counts = df["Estrellas"].value_counts().sort_index()

# Definir colores elegantes y suaves (sin naranja ni tonos fuertes)
colores_elegantes = ["#A7C7E7", "#98D7C2", "#FFD700", "#FFB6C1", "#B39EB5", "#6495ED"]  # Tonos azulados, dorados y pastel suaves

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    estrellas_counts, 
    labels=estrellas_counts.index.astype(int), 
    autopct='%1.1f%%', 
    colors=colores_elegantes, 
    startangle=140, 
    wedgeprops={"edgecolor": "white"},  # Bordes blancos para suavizar
    textprops={'fontsize': 12, 'color': "#4A4A4A"}  # Letras en gris oscuro para mayor legibilidad
)

# Agregar título con un color más neutro
plt.title("Distribución de Estrellas de los Alojamientos", fontsize=14, fontweight='bold', color='#4A4A4A')

# Agregar leyenda con etiquetas y colores
plt.legend(wedges, [f"{int(label)} Estrellas" for label in estrellas_counts.index], title="Número de Estrellas", loc="best", fontsize=11, frameon=False)

# Mostrar gráfico
plt.show()



````

**Interpretación:**  
* Más del 58% de los alojamientos no tiene una clasificación de estrellas oficial, lo que indica una gran presencia de apartamentos y alojamientos turísticos.
* Los hoteles de 4 y 5 estrellas representan el 18.7% y 6.2% respectivamente, lo
que sugiere que Madrid cuenta con una oferta significativamente de alojamientos
de lujo.

````python
# Importar librerías necesarias
import dash
from dash import dcc, html  # Importación corregida
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
import random

# Cargar los datos limpios
df = pd.read_csv("alojamientos_booking_limpio.csv")

# Convertir ubicación en coordenadas ficticias (si no tienes datos reales)
df["Latitud"] = df.index.map(lambda x: 40.40 + random.uniform(-0.1, 0.1))
df["Longitud"] = df.index.map(lambda x: -3.70 + random.uniform(-0.1, 0.1))

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Diseño del Dashboard
app.layout = html.Div(children=[

    html.H1(" Cuadro de Mando - Alojamientos en Madrid", style={'textAlign': 'center'}),

    # Selector de Distrito
    html.Label(" Selecciona un Distrito:"),
    dcc.Dropdown(
        id='distrito-dropdown',
        options=[{'label': 'Todos', 'value': 'Todos'}] + 
                [{'label': distrito, 'value': distrito} for distrito in df["Distrito"].unique()],
        value='Todos',
        clearable=False
    ),

    # Gráficos dinámicos
    dcc.Graph(id="mapa-hoteles"),
    dcc.Graph(id="ranking-distritos"),
    dcc.Graph(id="hist-precio"),
    dcc.Graph(id="scatter-precio-puntuacion"),
    dcc.Graph(id="top-reseñas"),
    dcc.Graph(id="pie-estrellas"),
    dcc.Graph(id="top-caros"),
    dcc.Graph(id="top-puntuacion"),
    dcc.Graph(id="top-estrellas"),
])

# Callback para actualizar gráficos dinámicamente
@app.callback(
    [Output("mapa-hoteles", "figure"),
     Output("ranking-distritos", "figure"),
     Output("hist-precio", "figure"),
     Output("scatter-precio-puntuacion", "figure"),
     Output("top-reseñas", "figure"),
     Output("pie-estrellas", "figure"),
     Output("top-caros", "figure"),
     Output("top-puntuacion", "figure"),
     Output("top-estrellas", "figure")],
    [Input("distrito-dropdown", "value")]
)
def update_graphs(distrito_seleccionado):
    # Filtrar por distrito
    df_filtrado = df if distrito_seleccionado == 'Todos' else df[df["Distrito"] == distrito_seleccionado]

    # 1 **Mapa Interactivo de Alojamientos**
    fig_mapa = px.scatter_mapbox(
        df_filtrado, lat="Latitud", lon="Longitud", hover_name="Alojamiento",
        hover_data=["Precio", "Puntuación", "Estrellas", "Reseñas"],
        color="Puntuación", size="Precio", size_max=15,
        title=" Ubicación de Alojamientos en Madrid",
        color_continuous_scale="viridis",
        zoom=12, height=500
    )
    fig_mapa.update_layout(mapbox_style="open-street-map")

    # 2 **Ranking de Distritos por Puntuación Media**
    df_ranking = df.groupby("Distrito")["Puntuación"].mean().reset_index().sort_values(by="Puntuación", ascending=False)
    fig_ranking_distritos = px.bar(
        df_ranking, x="Puntuación", y="Distrito", orientation='h',
        title=" Ranking de Distritos con Mejores Puntuaciones",
        color="Puntuación", color_continuous_scale="plasma",
        text_auto=True
    )

    # 3 **Histograma de Precios**
    fig_hist_precio = px.histogram(
        df_filtrado, x="Precio", nbins=20, title=" Distribución de Precios",
        color_discrete_sequence=["royalblue"]
    )

    # 4 **Scatter Plot - Relación entre Puntuación y Precio**
    fig_scatter_precio = px.scatter(
        df_filtrado, x="Puntuación", y="Precio",
        title=" Relación entre Puntuación y Precio",
        color="Puntuación", color_continuous_scale="viridis"
    )

    # 5 **Top 10 Alojamientos con Más Reseñas**
    top_reviews = df_filtrado.nlargest(10, "Reseñas")
    fig_top_reviews = px.bar(
        top_reviews, x="Reseñas", y="Alojamiento", orientation='h',
        title=" Top 10 Alojamientos con Más Reseñas",
        color="Reseñas", color_continuous_scale="magma",
        text_auto=True
    )

    # 6 **Gráfico de Pastel - Distribución de Estrellas**
    fig_pie_estrellas = px.pie(
        df_filtrado, names="Estrellas", title=" Distribución de Estrellas",
        color_discrete_sequence=px.colors.sequential.Blues
    )

    # 7 **Top 10 Hoteles Más Caros**
    top_caros = df_filtrado.nlargest(10, "Precio")
    fig_top_caros = px.bar(
        top_caros, x="Precio", y="Alojamiento", orientation='h',
        title=" Top 10 Hoteles Más Caros",
        color="Precio", color_continuous_scale="reds",
        text_auto=True
    )

    # 8 **Top 10 Hoteles con Mejor Puntuación**
    top_puntuacion = df_filtrado.nlargest(10, "Puntuación")
    fig_top_puntuacion = px.bar(
        top_puntuacion, x="Puntuación", y="Alojamiento", orientation='h',
        title=" Top 10 Hoteles con Mejor Puntuación",
        color="Puntuación", color_continuous_scale="teal",
        text_auto=True
    )

    # 9 **Top 10 Hoteles con Más Estrellas**
    top_estrellas = df_filtrado.nlargest(min(10, len(df_filtrado)), "Estrellas")
    fig_top_estrellas = px.bar(
        top_estrellas, x="Estrellas", y="Alojamiento", orientation='h',
        title=" Top 10 Hoteles con Más Estrellas",
        color="Estrellas", color_continuous_scale="sunsetdark",
        text_auto=True
    )

    return fig_mapa, fig_ranking_distritos, fig_hist_precio, fig_scatter_precio, fig_top_reviews, fig_pie_estrellas, fig_top_caros, fig_top_puntuacion, fig_top_estrellas

# Ejecutar la aplicación en un puerto diferente
if __name__ == '__main__':
    app.run_server(debug=True, port=8080)  # Cambia 8080 por el puerto que prefieras
````
